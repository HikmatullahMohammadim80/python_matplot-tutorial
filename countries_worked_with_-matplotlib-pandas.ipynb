```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
df_country = pd.read_csv('countries.csv')
```

```python
df_country.head()
```

```python
# take countries and populatoin and order them descendingly
ctry_pop = df_country[['Country', 'Population']].sort_values('Population')[::-1]
```

```python
type(ctry_pop)
```

```python
# 20 most populated countries
plt.style.use('bmh')
plt.barh(ctry_pop['Country'][:20][::-1], ctry_pop['Population'][:20][::-1])
plt.title('20 Most Poplulated Countries')
plt.xlabel('Population')
plt.show()
```

```python
# 20 least populated countries
plt.style.use('bmh')
plt.barh(ctry_pop['Country'][::-1][:20], ctry_pop['Population'][::-1][:20], color='green')
plt.title('20 Least Poplulated Countries')
plt.xlabel('Population')
plt.show()
```

```python
df_country.head()
```

```python
# Population according to regions
region_pop = df_country.groupby('Region')[['Population']]\
    .sum()\
    .reset_index()\
    .sort_values('Population')
region_pop
```

```python
plt.style.use('fast')
plt.barh(region_pop['Region'], region_pop['Population'])

plt.title('Population According to Regions')
plt.xlabel('Population')
plt.show()
```

```python
## 20 widest countries
ctry_lit =  df_country[['Country', 'Area (sq. mi.)']].sort_values('Area (sq. mi.)')
```

```python
x = ctry_lit['Area (sq. mi.)'].tail(20)
y = ctry_lit['Country'].tail(20)

plt.style.use('fast')
plt.barh(y, x)
plt.title('20 Widest Countries')
plt.xlabel('Area (sq. mi.)')
plt.show()
```

```python
## 20 least wide countries
ctry_lit =  df_country[['Country', 'Area (sq. mi.)']].sort_values('Area (sq. mi.)')
```

```python
x = ctry_lit['Area (sq. mi.)'].head(20)[::-1]
y = ctry_lit['Country'].head(20)[::-1]

plt.barh(y, x, color='green')
plt.title('20 Least Wide Countries')
plt.xlabel('Area (sq. mi.)')
plt.show()
```

```python
lit_region = df_country\
            .groupby('Region')['Literacy (%)']\
            .mean()\
            .reset_index()\
            .sort_values('Literacy (%)')
```

```python
# plot the data
plt.style.use('fast')
plt.barh(lit_region['Region'], lit_region['Literacy (%)'])
plt.title('Literacy According to Regions')
plt.grid(True)
plt.xlabel('Literacy (%)')
plt.show()
```

```python
# 20 least literate countries
```

```python
lit_ctry = df_country[['Country', 'Literacy (%)']]\
            .sort_values('Literacy (%)')

least = lit_ctry[:20][::-1]

plt.style.use('fast')
plt.barh(least['Country'], least['Literacy (%)'])
plt.grid(True)
plt.title('20 Least Literate Countries')
plt.xlabel('Literacy (%)')
plt.show()
```
